```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```



# **An√°lise de Correla√ß√£o entre Total da Conta e Gorjeta com Seaborn**


Um restaurante deseja investigar a rela√ß√£o entre o total da conta e o valor da gorjeta, segmentando os dados por g√™nero dos clientes. A proposta √© entender como o g√™nero pode influenciar o valor da gorjeta em fun√ß√£o do total da conta.




```python
# Carregar o dataset - 'tips' a fonte de dados sobre gorjetas
df = sns.load_dataset('tips')

# Criar a figura e os eixos para o gr√°fico, com 8px de largura e 4px de altura
fig, ax = plt.subplots(figsize=(8, 4))

# Criando o gr√°fico de dispers√£o (scatterplot) com Seaborn
sns.scatterplot(
    data=df, # Vari√°vel "data" referindo-se a dados que cont√©m o "DF", fonte de dados
    x='total_bill',  # Define a vari√°vel do eixo X (total da conta)
    y='tip',  # Define a vari√°vel do eixo Y (gorjeta)
    hue='sex',  # Concatena os pontos por sexo (diferencia entre 'Male' e 'Female')
    palette={'Male': 'blue', 'Female': 'pink'},  # Define as cores para cada sexo (azul para masculino, rosa para feminino)
)

# Adicionando uma linha de regress√£o ao gr√°fico de dispers√£o
sns.regplot(
    data=df,
    x='total_bill',  # Define a vari√°vel do eixo X.
    y='tip',  # Define a vari√°vel do eixo Y.
    scatter=False,  # N√£o plota os pontos de dispers√£o, apenas a linha de regress√£o
    color='black'  # Define a cor da linha de regress√£o como preta
)

# Ajustando t√≠tulo, legendas e estilos do gr√°fico
plt.title('Rela√ß√£o entre Gorjetas e Total da Conta por G√™nero', fontsize=16, fontweight='bold')  # Define o t√≠tulo do gr√°fico
plt.xlabel('Total da Conta (em $)', fontsize=12)  # Define o r√≥tulo do eixo X com a fonte de tamanho 12
plt.ylabel('Gorjeta (em $)', fontsize=12)  # Define o r√≥tulo do eixo Y com a fonte de tamanho 12
plt.legend(title='Sexo', fontsize=10, title_fontsize=12)  # Adiciona a legenda para o sexo, com t√≠tulos e fontes ajustadas
plt.grid(visible=True, linestyle='--', alpha=0.5)  # Adiciona uma grade no fundo do gr√°fico, com estilo de linha tracejada e opacidade de 50%

# Melhorando a est√©tica ajustando o tamanho da fonte dos ticks
plt.xticks(fontsize=10)  # Ajusta o tamanho da fonte para os r√≥tulos do eixo X
plt.yticks(fontsize=10)  # Ajusta o tamanho da fonte para os r√≥tulos do eixo Y

# Ajusta a distribui√ß√£o do layout, garantindo que os elementos n√£o fiquem sobrepostos
plt.tight_layout()
```
